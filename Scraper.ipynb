```python

# Written by Omer Khan lmao 

# ______READ THIS PLEASE ________
# some baisc instructions on how to use this 

# create two text files, myfile.txt and data.txt

# enter a url for a product page 
# it will auto scroll to the reviews and read them the first time
# for it to read more you have to clcik on the next button in the reviews section
# after about 8 pages, you want to scroll to the reviews and stay there while it writes them down to a file. 
# Dont scroll further, it might present you with an error 

# this code will save mostly english reviews, but its a good idea to look through your saved text just in case. 
# Also it saves unique reviews, duplicates are removed

# ___________________________________

# pip install these libraries 
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time 
from langdetect import detect


# get the chrome driver here https://sites.google.com/chromium.org/driver/downloads
# check your chrome version too before downloading 
# download the chrome driver online and give its folder path here 

# my computer is a mac, so windows users do your own thing 
chromedriver_path= "/Users/admin/Downloads/chromedriver_mac64/chromedriver"


options = Options()
options.add_argument("--verbose")
# options.add_argument("--headless")
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

URL= input("Please enter a url:")
driver = webdriver.Chrome(options=options, executable_path=chromedriver_path)
# You can manually use this by using the line below 
# driver.get("https://www.daraz.pk/products/hot-best-selling-mini-small-leather-wallet-and-card-holder-stylish-slim-wallet-for-men-boys-brown-blue-camel-chocolate-i314509574-s1820724512.html?&search=pdp_v2v?spm=a2a0e.pdp.recommendation_1.5.df6cd8wPd8wPC5&mp=1&scm=1007.38553.252219.0&clickTrackInfo=af8bc2a4-956c-4a99-aa3a-826aa7aa0efb__314509574__10002040__cate__330312__0.483593__0.483593__0.0__0.0__0.0__0.0174844__4__null__null__null__null__null__null____150.0__0.19999999999999996__4.53147__143__120.0__153780,272585,272932__null__null__null__3650.16544_955.3632_4559.21183__null__28556__null__0.0__0.0________null__null")

driver.get(URL)
html = driver.find_element(By.TAG_NAME, 'html')

# this lets the page load, and also moves the page down so the reviews are loaded. 
# if your internet is faster you can reduce the time.sleep() time

x=0
while x<6: 
    time.sleep(.5)
    html.send_keys(Keys.PAGE_DOWN)
    x=x+1
    

# daraz loads reviews asynchronusly, ONLY FIVE AT A TIME LIKE WHAT EVEN
# its a huge pain because look at that atrocious class name for the button, it prevents me from being able to automate it 
# However I did find a crappy work around HAHAHA 
# So all you have to do is click on the next button and all the newly loaded reviews will be read. You have 10 seconds to click on the button to load more


# raw data will be written to myfile.txt, while filtered reviews will be written to data.txt
file1 = open("myfile.txt", "a")  # append mode

empty=""
# this will write all the reviews it picks up, regardless of duplicates
z=0
while z<10:
    for element in driver.find_elements_by_class_name('content'):
        TEXT= element.text
        if (len(TEXT)>0):
            if (detect(TEXT)=='en'): 
                file1.write(element.text+"\n")
                print (element.text)
        
    # driver.find_element_by_class_name('next-btn next-btn-normal next-btn-medium next-pagination-item next').click()
    # driver.find_element_by_xpath("//button[@class='next-btn_next-btn-normal_next-btn-medium next-pagination-item_next']").click()
    time.sleep(10)
    print (z)
    z=z+1

file1.close()


# removes duplicates and saves to data.txt
lines_seen = set() # holds lines already seen
outfile = open("data.txt", "a")
for line in open("myfile.txt", "r"):
    if line not in lines_seen: # not a duplicate
        outfile.write(line)
        lines_seen.add(line)
outfile.close()
```
